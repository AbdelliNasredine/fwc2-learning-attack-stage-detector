# Data loading & clearning

```python
import sys

sys.path.append('../')

from fwc2.utils import load_data, clearn_data

datasets = ['dapt20', 'scvic21']
# datasets = ['unraveled']
data = []

for dataset in datasets:
    df = clearn_data(load_data(dataset))
    df.to_csv(f'../datasets/{dataset}/clean.csv', encoding='utf-8', index=False)
```

```python
for d in data:
    normal_data_mask = d['label'] == 'benign'
    
    normal_ratio = len(d[normal_data_mask]) / len(d)
    attk_ratio = len(d[~normal_data_mask]) / len(d)
    
    print(f'normal/attack rations = {normal_ratio * 100:.2f} , {attk_ratio * 100:.2f}')
```
