```python
import os
import pandas as pd

base_dir = '../datasets/unveald'
```

```python
columns = []
total_size = 0
dfs = []
for dir in os.listdir(base_dir):
    for filename in os.listdir(os.path.join(base_dir, dir)):
        df = pd.read_csv(os.path.join(base_dir, dir, filename), on_bad_lines='skip')
        total_size += df.shape[0]
        if len(columns) == 0:
            columns = df.columns
        dfs.append(df.copy())
total_size, columns
```

```python
all_df = pd.concat(dfs, ignore_index=True)
```

```python
all_df.shape
```

```python
all_df.to_csv(os.path.join(base_dir, 'all.csv'), encoding='utf-8')
```

```python
cleaned_df = pd.read_csv(os.path.join(base_dir, 'cleaned.csv'))
```

```python
# droping raws & coluns having all NULL values
print(cleaned_df.shape)

cleaned_df = cleaned_df.dropna(how='all', axis=0)
cleaned_df = cleaned_df.dropna(how='all', axis=1)

cleaned_df.shape
```

```python
cleaned_df['Stage'].value_counts()
```

```python
for col in cleaned_df.columns:
    print(f'{col} = {cleaned_df[col].isna().sum()}')
```

```python
# droping non changing columns, ie having zero variance / std
for col in cleaned_df.select_dtypes(exclude='object').columns:
    var = cleaned_df[col].var()
    print(f'{col} = {var}')

    if var == 0.0:
        cleaned_df = cleaned_df.drop(columns=[col])

cleaned_df.shape
```

```python
# keeping on stage as target variable and droping other lable cols
cleaned_df = cleaned_df.drop(columns=['Activity', 'DefenderResponse', 'Signature']) 
```

```python
cleaned_df['label'] = cleaned_df['Stage']

cleaned_df = cleaned_df.drop(columns=['Stage', 'Unnamed: 0'])
```
